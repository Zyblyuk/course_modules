### Описание признаков
##### Демографические данные

    CustomerID: Уникальный идентификатор клиента.
    BirthYear: Год рождения клиента.
    EducationLevel: Уровень образования клиента.
    MaritalStatus: Семейное положение клиента.
    AnnualIncome: Ежегодный доход домохозяйства клиента.
    NumChildren: Количество детей в семье клиента.
    NumTeenagers: Количество подростков в семье клиента.
    EnrollmentDate: Дата регистрации клиента в компании.

##### Поведение

    DaysSinceLastPurchase: Количество дней с момента последней покупки клиента.
    FiledComplaint: Указано, подавал ли клиент жалобы за последние 2 года (1 — да, 0 — нет).

##### Траты на продукты

    SpentOnWines: Сумма, потраченная на покупку вина за последние 2 года.
    SpentOnFruits: Сумма, потраченная на покупку фруктов за последние 2 года.
    SpentOnMeat: Сумма, потраченная на покупку мяса за последние 2 года.
    SpentOnFish: Сумма, потраченная на покупку рыбы за последние 2 года.
    SpentOnSweets: Сумма, потраченная на покупку сладостей за последние 2 года.
    SpentOnGold: Сумма, потраченная на покупку золота за последние 2 года.

##### Маркетинговые кампании

    DiscountedPurchases: Количество покупок со скидкой.
    AcceptedCampaign1: Указано, принял ли клиент предложение первой кампании (1 — да, 0 — нет).
    AcceptedCampaign2: Указано, принял ли клиент предложение второй кампании (1 — да, 0 — нет).
    AcceptedCampaign3: Указано, принял ли клиент предложение третьей кампании (1 — да, 0 — нет).
    AcceptedCampaign4: Указано, принял ли клиент предложение четвертой кампании (1 — да, 0 — нет).
    AcceptedCampaign5: Указано, принял ли клиент предложение пятой кампании (1 — да, 0 — нет).
    LastCampaignResponse: Указано, принял ли клиент предложение последней кампании (1 — да, 0 — нет).

##### Каналы покупок

    OnlinePurchases: Количество покупок, совершенных через веб-сайт компании.
    CatalogPurchases: Количество покупок, совершенных через каталог.
    InStorePurchases: Количество покупок, совершенных в магазине.
    WebsiteVisitsLastMonth: Количество визитов на веб-сайт компании за последний месяц.

##### Финансовые метрики

    ContactCost: Стоимость контакта с клиентом (постоянная метрика).
    RevenueGenerated: Доход, сгенерированный клиентом (постоянная метрика).

```python
columns_to_drop = ['CustomerID', 'ContactCost', 'RevenueGenerated']
df = df.drop(columns=columns_to_drop, errors='ignore')

df['EnrollmentDate'] = pd.to_datetime(df['EnrollmentDate'], format='%d-%m-%Y')

median_income = df['AnnualIncome'].median()
df['AnnualIncome'] = df['AnnualIncome'].fillna(median_income)

missing_values = df.isnull().sum()

missing_values[missing_values > 0]
```

```python
df['EducationLevel'] = df['EducationLevel'].replace(
    {'Basic': 'Undergraduate', 
     'Graduation': 'Graduate', 
     'Master': 'Postgraduate', 
     '2n Cycle': 'Postgraduate', 
     'PhD': 'Postgraduate'}
)

df['MaritalStatus'] = df['MaritalStatus'].replace(
    {'Married': 'Partner', 'Together': 'Partner',
     'Single': 'Single', 'Divorced': 'Single', 'Widow': 'Single', 
     'Alone': 'Single', 'Absurd': 'Single', 'YOLO': 'Single'}
)

cols_to_corr = ['BirthYear', 'AnnualIncome', 'NumChildren', 'NumTeenagers', 'SpentOnWines', 'SpentOnMeat']

plt.figure(figsize=(8, 6))
corr_matrix = df[cols_to_corr].corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.show()
```

```python
df['CustomerAge'] = 2015 - df['BirthYear']
df['TotalChildren'] = df['NumChildren'] + df['NumTeenagers']
df['FamilySize'] = df['TotalChildren'] + df['MaritalStatus'].apply(lambda x: 2 if x == 'Partner' else 1)

df['TotalSpent'] = (df['SpentOnWines'] + df['SpentOnFruits'] + 
                    df['SpentOnMeat'] + df['SpentOnFish'] + 
                    df['SpentOnSweets'] + df['SpentOnGold'])

df['TotalPurchases'] = df['OnlinePurchases'] + df['CatalogPurchases'] + df['InStorePurchases']

max_date = df['EnrollmentDate'].max()
df['CustomerTenure'] = (max_date - df['EnrollmentDate']).dt.days

df['Recency'] = df['DaysSinceLastPurchase']
df['Frequency'] = df['TotalPurchases']
df['Monetary'] = df['TotalSpent']

df['OnlineShare'] = df['OnlinePurchases'] / df['TotalPurchases']
df['CatalogShare'] = df['CatalogPurchases'] / df['TotalPurchases']
df['StoreShare'] = df['InStorePurchases'] / df['TotalPurchases']

df[['OnlineShare','CatalogShare','StoreShare']] = df[['OnlineShare','CatalogShare','StoreShare']].fillna(0)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

cols_to_drop = ['BirthYear', 'EnrollmentDate', 'NumChildren', 'NumTeenagers'] # тк новые создали
df_model = df.drop(columns=cols_to_drop)

df_model = pd.get_dummies(df_model, columns=['EducationLevel', 'MaritalStatus'])

features_for_clustering = [
    'AnnualIncome',
    'CustomerAge',
    'FamilySize',
    'Recency',
    'Frequency',
    'Monetary',
    'OnlineShare',
    'CatalogShare',
    'StoreShare'
]

# Убрал категориальные признаки тк из-за них были оч хороший результат и-за неверного расчета растояний
# cat_cols = [col for col in df_model.columns if 'EducationLevel_' in col or 'MaritalStatus_' in col]
# features_for_clustering.extend(cat_cols)

X = df_model[features_for_clustering]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

pca = PCA(n_components=3, random_state=42)  # тестировал, вышел лучший результат
X_final = pca.fit_transform(X_scaled_df)
print(X_final)
# display(X_final.head())
```

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_final)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--', color='b')
plt.xlabel('n_clusters')
plt.ylabel('WCSS')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

sil_scores = {}

for k in range(2, 15):
    clust = AgglomerativeClustering(n_clusters=k, linkage="average", metric="cosine")
    labels = clust.fit_predict(X_final)
    sil_scores[k] = silhouette_score(X_final, labels)

plt.figure(figsize=(8,5))
plt.plot(range(2,15), sil_scores.values(), marker='o')
plt.xlabel("n_clusters")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

sil_scores = {}

for k in range(2, 15):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_final)
    sil_scores[k] = silhouette_score(X_final, labels)

plt.figure(figsize=(8,5))
plt.plot(range(2,15), sil_scores.values(), marker='o')
plt.xlabel("n_clusters")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.show()
```

```python

best_cluster_count = 4 # оптимальное значение
```

```python
from sklearn.metrics import silhouette_score

kmeans = KMeans(n_clusters=best_cluster_count, init='k-means++', random_state=42)
cluster_labels = kmeans.fit_predict(X_final)

sil_score = silhouette_score(X_final, cluster_labels)
print(f"KMeans (Silhouette Score): {sil_score}")

df['Cluster_kmeans'] = cluster_labels

print(df['Cluster_kmeans'].value_counts().sort_index())
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

linkages = ['ward', 'average', 'complete']
metrics = ['euclidean', 'cosine', 'manhattan']

best_score = 0
best_model = None
best_linkage = ""
best_metric = ""

print(f"best case for AgglomerativeClustering (k={best_cluster_count}):\n")

for link in linkages:
    valid_metrics = ['euclidean'] if link == 'ward' else metrics
    
    for metric in valid_metrics:
        agg = AgglomerativeClustering(n_clusters=best_cluster_count, metric=metric, linkage=link)
        labels = agg.fit_predict(X_final)
        score = silhouette_score(X_final, labels, metric=metric)
        
        print(f"Linkage: {link} | Metric: {metric} | Silhouette: {score}")
        
        if score > best_score:
            best_score = score
            best_model = agg
            best_linkage = link
            best_metric = metric

df['Cluster'] = best_model.labels_


print(f"Silhouette Score: {best_score:.4f}")
```

```python
cluster_profile = df.groupby('Cluster').agg({
    'CustomerAge': 'mean',
    'AnnualIncome': 'mean',
    'TotalChildren': 'mean',
    'TotalSpent': 'mean',
    'TotalPurchases': 'mean',
    'SpentOnWines': 'mean',
    'SpentOnMeat': 'mean',
    'DiscountedPurchases': 'mean',
    'AcceptedCampaign1': 'mean',
    'AcceptedCampaign5': 'mean',
    'LastCampaignResponse': 'mean'
}).round(2)

display(cluster_profile.T)
```
