# 14. Какая команда выиграла наибольшое количество матчей в сезоне? (Будьте внимательны с типом матча в сезоне - season).

Я сложил все домашние и гостевые победы команд по каждому сезону и нашёл того, у кого общее число выигрышей оказалось самым большим.

```python
home_wins = df.groupby(["season", "home_team"])["home_team_win"].sum().rename_axis(index={"home_team": "team"})
away_wins = df.groupby(["season", "away_team"])["home_team_loss"].sum().rename_axis(index={"away_team": "team"})

total_wins = home_wins + away_wins
total_wins.groupby("team").sum().idxmax()
```

```python
home_wins.loc[home_wins.groupby("season").idxmax()]
```

# 17. Какая команда проиграла наибольшее количество матчей в сезоне?

Суммировал поражения как дома, так и в гостях.

```python
home_lost = df.groupby(["season", "home_team"])["home_team_loss"].sum().rename_axis(index={"home_team": "team"})
away_lost = df.groupby(["season", "away_team"])["home_team_win"].sum().rename_axis(index={"away_team": "team"})

total_lost = home_lost + away_lost
total_lost.loc[total_lost.groupby("season").idxmax()]
```
